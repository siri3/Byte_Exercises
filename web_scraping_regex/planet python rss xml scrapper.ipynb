XML Data Format Excercise

Parse the XML data from the below link
URL http://planetpython.org/rss20.xml

and present the data under items channel/item
title, pubDate, and link


```python
import requests
```

```python
import xml.etree.ElementTree as ET
```

```python
url = 'http://planetpython.org/rss20.xml'
```

```python
r = requests.get(url)
```

```python
r
```

```python
content = r.content
```

```python
content
```

```python
decoded = content.decode()
```

```python
decoded
```

```python
type(decoded)
```

```python
#tree = ET.parse(decoded)   #file name too long error
```

```python
with open ("pp_rss.xml","w") as file:
    file.writelines(decoded)
```

```python
tree = ET.parse("pp_rss.xml")
```

```python
tree
```

```python
root = tree.getroot()  
```

```python
root
```

```python
root.tag
```

```python
root.attrib
```

```python
print (root.tag,root.text)
channel = root[0]
print (channel.tag,channel.text) #text has new line or tab
for items in channel:
    print (items.tag,items.text) 
    for child in items:
            print (items.tag,child.tag,child.text)
    
```

```python
channel = root.find('channel')
```

```python
channel
```

```python
type(channel)
```

```python
items = channel.findall('item')
```

```python
for elem in items:
    print(elem.findtext('title'))
    print(elem.findtext('link'))
    print(elem.findtext('pubDate'))
```
